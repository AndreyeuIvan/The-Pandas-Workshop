```python
#
# libraries
#
import pandas as pd
```

```python
#
# read in data
#
retail_sales = pd.read_csv('../datasets/online_retail_II.csv')
retail_sales.head()
```

```python
#
# convert InvoiceDate from str to datetime
# inspect the earliest and latest datetimes
#
retail_sales['InvoiceDate'] = pd.to_datetime(retail_sales['InvoiceDate'])
print('start: ', retail_sales.InvoiceDate.min(), 
      '\nend: ', retail_sales.InvoiceDate.max())
```

```python
#
# convert the dates to the month ending dates
# drop the time component by selecting .dt.date
#
retail_sales['InvoiceDate'] = (retail_sales['InvoiceDate'] + 
                               pd.tseries.offsets.MonthEnd(0)).dt.date
retail_sales.head()
```

```python
#
# compute and save the revenue per line
#
retail_sales['Revenue'] = (retail_sales.loc[:, 'Quantity'] * 
                           retail_sales.loc[:, 'Price'])
#
# sum up revenue by month
#
sales_by_month = \
  retail_sales[['InvoiceDate', 'Revenue']].groupby('InvoiceDate').sum()
sales_by_month
```
