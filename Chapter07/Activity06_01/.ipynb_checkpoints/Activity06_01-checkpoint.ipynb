```python
#2 Import the pandas package:
import pandas as pd
```

```python
#3 Load the CSV file as a DataFrame:
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter06/Data/titanic.csv'
data_frame = pd.read_csv(file_url)
```

```python
#4 Display the first 10 rows of the DataFrame:
data_frame.head(10)
```

```python
#5 Display rows with missing data:
data_frame[data_frame.isnull().any(1)]
```

```python
#6 Remove rows with missing data then display the DataFrame:
data_frame = data_frame.dropna()
data_frame
```

```python
#7 Build a pivot table indexed on "gender":
pd.pivot_table(data_frame,index=['gender'])
```

```python
#8 Build a pivot table indexed on "gender" and "ticket_class":
pd.pivot_table(data_frame,index=['gender','ticket_class'])
```

```python
#9 Build a pivot table indexed on "gender" and "ticket_class" with the mean and su aggregation on the "survived" column:
pd.pivot_table(data_frame,index=['gender','age_group','ticket_class'], values = 'survived',aggfunc={'mean','sum'})
```
