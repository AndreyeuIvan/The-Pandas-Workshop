```python
#2 Import pandas:
import pandas as pd
```

```python
#3 Next, load the CSV file as a DataFrame with the header parameter set to None :
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter06/Data/messy_addresses.csv'
data_frame = pd.read_csv(file_url, header=None)
```

```python
#4 Use the head() function to display the first 5 rows of the DataFrame:
data_frame.head()
```

```python
#5 Use the tail() function to display the last 5 rows of the DataFrame:
data_frame.tail(5)
```

```python
#6 Delete the duplicate columns then display the DataFrame
data_frame.drop([2,3], axis=1, inplace=True)
data_frame
```

```python
#7 Create list of column headers
column_names = ["full_name", "address"]
```

```python
#8 Replace the missing column headers then display the DataFrame
data_frame.columns = column_names
data_frame
```

```python
#9 Split the values of “full_name” into 2 columns called “first_name” and “last_name”) then display the DataFrame:
data_frame[['first_name','last_name']]=data_frame.full_name.str.split(expand=True)
data_frame
```

```python
#10 Delete the "full_name" column then display the DataFrame:
data_frame.drop('full_name', axis=1, inplace=True)
data_frame
```

```python
#11 Split the values of "address" into 3 new columns called "street", "city","state" then display the DataFrame:
data_frame[['street', 'city','state']] = data_frame.address.str.split(pat = ", ", expand=True)
data_frame
```

```python
#12 Delete the "address" column then display the DataFrame:
data_frame.drop('address', axis=1, inplace=True)
data_frame
```
