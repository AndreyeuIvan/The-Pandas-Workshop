```python
#
# required libraries
#
import pandas as pd
#
```

```python
#
# read in the file mushroom.csv
#
soybean_diseases = pd.read_csv('..\\Datasets\\soybean.csv')
soybean_diseases
#
```

```python
#
# select the condition column, and display how many unique conditions there are
# print the result out in a human readable format
#
print('there are', len(soybean_diseases['condition'].unique()), 
      'unique conditions', 
      [soybean_diseases['condition'].unique()[i]
       for i in range(len(soybean_diseases['condition'].unique()))])
#
```

```python
#
# find all the cases of "brown-spot" and print out the number
#
brown_spots = \
    soybean_diseases.loc[soybean_diseases['condition'] == 'brown-spot', :]
print('there are', brown_spots.shape[0], 'instances having brown-spot')
#
```

```python
#
# how many conditions are associated with hail damage (hail == 1)
# out of how many total conditions?
#
hail_related = \
    soybean_diseases.loc[soybean_diseases['hail'] == 1, 'condition'].unique()
print('there are', len(hail_related), 
      'conditions associated with hail damage out of',
      len(soybean_diseases['condition'].unique()), 
      'total conditions')
#
```

```python
#
# create a new DataFrame with only the hail-related entries
# print the number of cases compared to the original total
#
hail_cases = soybean_diseases.loc[soybean_diseases['hail'] == 1, :]
print('there are ', hail_cases.shape[0], ' hail-related cases out of ',
      soybean_diseases.shape[0], ' total cases')
#
```

```python
#
# loop over hail cases, and print all conditions with most severe damage (severity == 2)
#
for i in range(hail_cases.shape[0]):
    if hail_cases.loc[i, 'severity'] == 2:
        print('case ', i, ' with condition ', 
              hail_cases.loc[i, 'condition'], ' is severe')
#
```

```python
#
# inspect the index of hail_cases
#
hail_cases.index
```

```python
#
# fix the problem by iterating over the index instead of integers
#
for i in hail_cases.index:
    if hail_cases.loc[i, 'severity'] == 2:
        print('case ', i, ' with condition ', 
              hail_cases.loc[i, 'condition'], ' is severe')
```
