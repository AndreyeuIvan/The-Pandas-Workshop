```python
#
# required libraries
#
import pandas as pd
#
```

```python
#
# read in the file mushroom.csv
#
mushroom = pd.read_csv('Datasets\\mushroom.csv')
mushroom
#
```

```python
#
# remove the poisonous mushrooms from the data
#
mushroom = mushroom.loc[mushroom['class'] == 'e', :]
mushroom.shape
#
```

```python
#
# try printing all the classes using a loop
#
for i in range(mushroom.shape[0]):
    print(mushroom['class'][i])
#
```

```python
#
# display the first 25 values of the index
#
list(mushroom.index)[:25]
#
```

```python
#
# print the first 25 class values
#
# method 1
#
print('method 1--use the index directly\n')
for i in list(mushroom.index)[:25]:
    print(mushroom['class'][i])
#
# method 2
#
print('method 2--reset the index\n')
mushroom.reset_index(drop = True, inplace = True)
for i in range(25):
    print(mushroom['class'][i])
#
```
