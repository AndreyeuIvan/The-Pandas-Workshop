```python
#
# create a DataFrame to show integer index values vs. labels
#
labels_vs_integers = pd.DataFrame({'values' : [6, 1, 5, 2],
                                  'names' : ['oranges', 'apples', 
                                             'bananas', 'pears']})
labels_vs_integers
```

```python
#
# show the names in the non_citrus rows using
# both .iloc and .loc
#
print(labels_vs_integers.iloc[1:4, 1])
print()
print(labels_vs_integers.loc['non_citrus', 'names'])
```

```python
species = pd.Series(['cat', 'dog', 'pig', 'chicken']).sample(100, replace = True, random_state = 1).reset_index(drop = True)
location = pd.Series(['city', 'town', 'farm']).sample(100, replace = True, random_state = 2).reset_index(drop = True)
weight = pd.Series(range(10, 110)).reset_index(drop = True)
fur = pd.Series(['long', 'short']).sample(100, replace = True, random_state = 3).reset_index(drop = True)
color = pd.Series(['solid', 'spotted', 'striped']).sample(100, replace = True, random_state = 42).reset_index(drop = True)
animals = pd.DataFrame({'species' : species,
                        'location' : location,
                        'weight' : weight,
                        'color' : color,
                        'fur' : fur})
animals
```

```python
pd.options.display.float_format = '{:,.2f}'.format
animals.sort_values(['location', 
                     'species']).pivot_table(index = ['location', 'species'],
                                             columns = ['color', 'fur'],
                                             aggfunc = 'mean',
                                             fill_value = '')
```

```python
#
# read in the data without the multiindex
#
GDP_by_industry = pd.read_csv('Datasets\\US_GDP_Industry.csv')
GDP_by_industry.head()
```

```python
#
# create the multi-index
#
GDP_index = pd.MultiIndex.from_frame(GDP_by_industry[['period', 'Industry']])
GDP_index
```

```python
#
# replace the index with the multi-index
#
# drop the columns first
#
GDP_by_industry.drop(columns = ['period', 'Industry'], inplace = True)
GDP_by_industry.set_index(GDP_index, drop = True, inplace = True)
GDP_by_industry
```

```python
#
# sort the data 
#
GDP_by_industry.sort_values(by = ['period', 'Industry'], inplace = True)
GDP_by_industry
```
