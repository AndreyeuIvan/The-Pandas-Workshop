```python
#2 Import pandas:
import pandas as pd
```

```python
#3 Next, load the CSV file as a DataFrame on the notebook and read the CSV file:
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter04/Data/retail_purchase_missing.csv'
data_frame = pd.read_csv(file_url)
```

```python
#4 Use the head() function to display the first 5 rows of the DataFrame:
data_frame.head()
```

```python
#5 Use the info() function to display the data types of each column in the DataFrame:
data_frame.info()
```

```python
#6 Replace the missing values by 0 in every numeric columns by using fillna() then display the top 5 rows:
data_frame.fillna(value = {'Receipt Id': 0, 'Product Weight': 0, 'Total Price': 0}, inplace= True)
data_frame.head()
```

```python
#7 Replace the missing values by '01/01/99' in “Date of Purchase”  by using fillna() then display the top 5 rows:
data_frame['Date of Purchase'].fillna('01/01/99',inplace = True)
data_frame.head()
```

```python
#8 Replace the missing values by 'Missing Name' in the remaining columns  by using fillna() then display the top 5 rows:
data_frame.fillna(value = {'Product Name': 'Missing Name', 'Retail shop name': 'Missing Name'}, inplace= True)
data_frame.head()
```

```python
#9 Convert the following columns then use info()to display the dtype of every columns :
data_frame['Date of Purchase'] = pd.to_datetime(data_frame['Date of Purchase'], format='%d/%m/%y')

# Convert 'Total Price' into a int
data_frame['Receipt Id'] = data_frame['Receipt Id'].astype('int')

# Convert 'Total Price' into a int
data_frame['Total Price'] = data_frame['Total Price'].astype('int')

# Convert 'Product Weight' into a int
data_frame['Product Weight'] = data_frame['Product Weight'].astype('int')

# Convert 'Product Name' into a category
data_frame['Product Name'] = data_frame['Product Name'].astype('category')

# Convert 'Retail shop name' into a category
data_frame['Retail shop name'] = data_frame['Retail shop name'].astype('category')

# Display data types of each columns of the DataFrame
data_frame.info()
```
