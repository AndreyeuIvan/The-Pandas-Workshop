```python
#
# required libraries
#
import pandas as pd
#
```

```python
#
# load data from a .csv file to a DataFrame
#
fname = 'Datasets/test_data_frame.csv'
economic_data = pd.read_csv(fname)
economic_data.head()
```

```python
#
# look at the data types of the data frame
#
economic_data.dtypes
```

```python
#
# convert the timestamps from object to a pandas timestamp
#
economic_data['time'] = pd.to_datetime(economic_data['time'],
                                      format = '%I:%M:%S %p')
economic_data['time'] = [economic_data['time'][i] - economic_data['time'][0]
                        for i in economic_data.index]
economic_data
```

```python
#
# now that the time column is a pandas time type
# let's replace the index with it
#
economic_data.set_index('time', drop = True, inplace = True)
economic_data.head()
```

```python
#
# now that we have a time-based index, we can apply some 
# powerful Pandas methods
# .resample() allows us to change the frequency of the
# observations, and applying .interpolate() fills the 
# new time slots with interpolated values
#
economic_data = economic_data.resample('5min').interpolate()
economic_data.head()
```

```python
#
# save the data to economic_data.csv
#
economic_data.to_csv('Datasets/economic_data.csv')
```
